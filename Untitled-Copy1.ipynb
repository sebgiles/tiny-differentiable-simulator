```python
def grad_plot(X, Y, Z11, Z12, Z21, Z22):
    max2 = np.max([Z21, Z22])
    min2 = np.min([Z21, Z22])
#     max2 = np.max([Z21])
#     min2 = np.min([Z21])
    cmap = plt.get_cmap()
    levelsY = MaxNLocator(nbins=cmap.N).tick_values(min2, max2)
    norm = BoundaryNorm(levelsY, ncolors=cmap.N, clip=True)

    plt.figure()
    
    ax = plt.subplot(2,2,1)
    plt.pcolormesh(X, Y, Z11, cmap=cmap)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('z')
    plt.colorbar()
    
    ax = plt.subplot(2,2,2)
    plt.pcolormesh(X, Y, Z12, cmap=cmap)
    plt.title('abs(diff-J)')
    plt.colorbar()    
    
    ax = plt.subplot(2,2,3)
    plt.pcolormesh(X, Y, Z21, cmap=cmap, norm=norm)
    plt.title('J')
    plt.colorbar()
    

    ax = plt.subplot(2,2,4)
#     plt.pcolormesh(X, Y, Z22, cmap=cmap)
    plt.pcolormesh(X, Y, Z22, cmap=cmap, norm=norm)
    plt.title('diff')
    plt.colorbar()
    
    plt.show()
```

```python
STATE_SIZE = 36
INPUT_SIZE = 12
STATE_INPUT_SIZE = STATE_SIZE + INPUT_SIZE
```

```python
import ocs2_whole_body_model_pybind as bindings
wb = bindings.WholeBody("/home/sgiles/anymal_ws/src/anymal_c_simple_description/urdf/anymal.urdf")
wb.generate_code(False)

params = bindings.WholeBody.Params()
params.maxIter = 8;     # This is frozen to 8 in generated code.
params.mu = 0.7;        # Contact friction coefficient
params.delta_t = 0.01;  # Time interval
params.eps_n = 0.00;    # Normal restitution coefficient
params.eps_t = 0.00;    # Tangential restitution coefficient
params.alpha = 0.8;     # prox convergence tuning factor
params.weightlessness = 0.00;
params.feedforward_weight = 0.0;

def model(xu):
    assert len(xu) == STATE_INPUT_SIZE
    x = xu[:STATE_SIZE]
    u = xu[STATE_SIZE:]
    xdot = wb.get_value(x, u, params)
    J = wb.get_jacobian(x, u, params)
    return xdot, J
```

```python
import pytinydiffsim_dual as pd
from pathlib import Path

anymal_urdf_path = Path("/home/sgiles/anymal_ws/src/anymal_c_simple_description/urdf/anymal2.urdf")
plane_urdf_path = Path("/home/sgiles/src/tiny-differentiable-simulator/data/plane_implicit.urdf")
assert anymal_urdf_path.exists()
assert plane_urdf_path.exists()

parser = pd.TinyUrdfParser()
urdf_data = parser.load_urdf(str(anymal_urdf_path))
plane_urdf_data = parser.load_urdf(str(plane_urdf_path))

plane_mb = pd.TinyMultiBody(False)
is_floating = True
mb = pd.TinyMultiBody(is_floating)

world = pd.TinyWorld()

pd.UrdfToMultiBody2().convert2(plane_urdf_data, world, plane_mb)
pd.UrdfToMultiBody2().convert2(urdf_data, world, mb)

mb_solver = pd.TinyMultiBodyConstraintSolver()

ts = pd.fraction(1, 100)
g = pd.Vector3(0, 0, -9.81)
mb_solver.cfm_ = pd.fraction(1, 500)
mb_solver.erp_ = pd.fraction(1, 5)
mb_solver.pgs_iterations_ = 50
mb_solver.keep_all_points_ = False
world.friction = pd.TinyDualDouble(1.8,0)

def analyse_model():
    link_ids = {l.joint_name: i for i, l in enumerate(mb.links)}
    link_ids['base'] = -1
    link_names = {v: k for k, v in link_ids.items()}
    cg = {k: mb.collision_geometries(v) for k, v in link_ids.items()}
    cg = {k: v for k, v in cg.items() if v}
    return link_ids, link_names, cg

def assign(lv, rv):
    for i in range(lv.size()):
        lv[i] = rv[i]

def step(state, action):
    assign(mb.q, state[:19])
    assign(mb.qd, state[19:])
    assign(mb.tau, action)
    pd.forward_kinematics(mb, mb.q, mb.qd)
    pd.forward_dynamics(mb, g)  # update qdd
    pd.integrate_euler_qdd(mb, ts)  # update qd, zero qdd
    dispatcher = world.get_collision_dispatcher()
    contacts = world.compute_contacts_multi_body([plane_mb, mb], dispatcher)[0]
    # we get 100 contacts:
    #   4:   1 for each sphere
    #   96:  8 for each box
    collisions = mb_solver.resolve_collision(contacts, ts)
    pd.integrate_euler(mb, ts)
    q_out = [mb.q[i] for i in range(mb.q.size())]
    qd_out = [mb.qd[i] for i in range(mb.qd.size())]
    return [*q_out, *qd_out]
```

```python
ZCOM = 2
TAU_KFE = STATE_SIZE + 2
DDOT_KFE = STATE_SIZE // 2 + 6 + 2
```

```python
X_IDX = ZCOM
Y_IDX = TAU_KFE
Z_IDX = DDOT_KFE
```

```python
def eval_1d(model, xu0, xu1, n_steps):
    xu_grid = np.linspace(xu0, xu1, n_steps+1)
    values = [model(xu) for xu in xu_grid]
    xdot_grid, J_grid = list(zip(*values))
    xdot_grid = np.array(xdot_grid)
    J_grid = np.array(J_grid)
    return xu_grid, xdot_grid, J_grid

# Return n evenly spaced points on an axis-aligned 2d grid in R^N
# a and b are diagonally opposite vertices of the grid
def grid_2d(a, b, n_x, n_y):
    assert a.shape == b.shape
    assert np.count_nonzero(a-b) == 2
    x = np.linspace(a[X_IDX], b[X_IDX], n_x)
    y = np.linspace(a[Y_IDX], b[Y_IDX], n_y)
    x_grid, y_grid = np.meshgrid(x, y, indexing='ij')
    xy_grid = np.array([x_grid.T, y_grid.T]).transpose([2,1,0])
    full_grid = np.tile(a, [n_x, n_y, 1])
    full_grid[:,:, [X_IDX, Y_IDX]] = xy_grid
    return full_grid

def eval_2d(model, xu0, xu1, n_xsteps, n_ysteps):
    n_x = n_xsteps + 1
    n_y = n_ysteps + 1
    xu_grid = grid_2d(xu0, xu1, n_x, n_y)    
    values = [model(xu) for row in xu_grid for xu in row]
    xdot_grid, J_grid = list(zip(*values))
    xdot_grid = np.array(xdot_grid).reshape([n_x, n_y, STATE_SIZE])
    J_grid = np.array(J_grid).reshape([n_x, n_y, STATE_SIZE, STATE_SIZE+INPUT_SIZE])
    return xu_grid, xdot_grid, J_grid

def compare_gradients_1d(model, xu0, xu1, n_steps):
    xu_grid, xdot_grid, J_grid = eval_1d(model, xu0, xu1, n_steps)
    print(J_grid.shape)
    x = xu_grid[:,X_IDX]
    y = xu_grid[:,Y_IDX]
    delta_xy = np.diff(np.stack([x,y]))
    z = xdot_grid[:,Z_IDX]
    J_xy = J_grid[:,Z_IDX,[X_IDX, Y_IDX]]
                  
    dz_fwd = (J_xy[:-1].T * delta_xy).sum(axis=0)
    dz_bwd = (J_xy[1:].T * delta_xy).sum(axis=0)
    dz = np.diff(z)
    plt.plot(x, z)
    plt.show()
    x = (x[:-1] + x[1:]) / 2
    plt.plot(x, dz_fwd)
    plt.plot(x, dz_bwd)
    plt.plot(x, dz)
    plt.legend(['fwd', 'bwd', 'diff'])
    plt.show()

def compare_gradients_2d(model, xu0, xu1, n_steps_x, n_steps_y):
    xu_grid, xdot_grid, J_grid = eval_2d(model, xu0, xu1, n_steps_x, n_steps_y)
    x = xu_grid[:, 0, X_IDX]
    y = xu_grid[0, :, Y_IDX]
    Z = xdot_grid[:, :, Z_IDX]
    dZdx = J_grid[:, :, Z_IDX, X_IDX]
    dZdy = J_grid[:, :, Z_IDX, Y_IDX]
    X = xu_grid[:, :, X_IDX]
    Y = xu_grid[:, :, Y_IDX]

    dX = np.diff(X, axis=0)
    dY = np.diff(Y, axis=1)

    dZ_fwdx = dZdx[1:, :] * dX
    # dZ_bwdx = dZdx[:-1, :] * dX
    dZ_fwdy = dZdy[:, 1:] * dY
    #dZ_bwdy = dZdy[:, :-1] * dY

    dZ_x = np.diff(Z, axis=0)
    dZ_y = np.diff(Z, axis=1)

    dZ_x_err = np.abs(dZ_x - dZ_fwdx)
    dZ_y_err = np.abs(dZ_y - dZ_fwdy)

    print('x:')
    grad_plot(X, Y, Z, dZ_x_err, dZ_fwdx, dZ_x)
    print('y:')
    grad_plot(X, Y, Z, dZ_y_err, dZ_fwdy, dZ_y)
```

```python
def model_tds(xu):
    assert len(xu) == STATE_INPUT_SIZE
    xu = list(xu)                             # allow heterogenous data
    xu.insert(3, 1.0)  # the input was euler angles
    xu[6] = xu[2]
    xu[2] = 0.0
    J = []
    for i, xui in enumerate(xu):
        xu[i] = pd.TinyDualDouble(xui, 1.0)   # replace i-th input with special object
        x = xu[:STATE_SIZE+1]
        u = xu[STATE_SIZE+1:]
        output = get_value(x, u)              # compute output and dual
        J.append([o.dual() for o in output])  # use dual to fill i-th column of jacobian
        xu[i] = xui                           # put back the normal value
    tmp = J[2]
    J[2] = J[6]
    J[6] = tmp
    tmp = J[2 + 18]
    J[2 + 18] = J[6 + 18]
    J[6 + 18] = tmp
    del J[3]
    xdot = [o.real() for o in output]
    tmp = xdot[2]
    xdot[2] = xdot[6]
    xdot[6] = tmp
    tmp = xdot[2 + 18]
    xdot[2 + 18] = xdot[6 + 18]
    xdot[6 + 18] = tmp
    return np.array(xdot), np.array(J).T
```

```python
ts2 = pd.fraction(1, 200)
def get_value(state, action):
    assign(mb.qd, state[19:])
    qd_out = [mb.qd[i] for i in range(mb.qd.size())]
    
    assign(mb.q, state[:19])
    pd.forward_kinematics(mb, mb.q, mb.qd)
    assign(mb.tau, action)
    pd.forward_dynamics(mb, g)  # update qdd
    
    # qdd_out = [mb.qdd[i] for i in range(mb.qdd.size())]
    
    pd.integrate_euler_qdd(mb, ts)  # update qd, zero qdd
    
    dispatcher = world.get_collision_dispatcher()
    contacts = world.compute_contacts_multi_body([plane_mb, mb], dispatcher)[0]
    collisions = mb_solver.resolve_collision(contacts, ts)  # update qd
    
    qdd_out = [(mb.qd[i] - qd_out[i]) / ts for i in range(mb.qd.size())]
    
    pd.integrate_euler(mb, ts)
    

    return [*qd_out, *qdd_out]
```
