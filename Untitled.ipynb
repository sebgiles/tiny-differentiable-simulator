```python
def analyse_model():
    link_ids = {l.joint_name: i for i, l in enumerate(mb.links)}
    link_ids['base'] = -1
    link_names = {v: k for k, v in link_ids.items()}
    cg = {k: mb.collision_geometries(v) for k, v in link_ids.items()}
    cg = {k: v for k, v in cg.items() if v}
    return link_ids, link_names, cg

def assign(lv, rv):
    for i in range(lv.size()):
        lv[i] = rv[i]

def step(state, action):
    assign(mb.q, state[:19])
    assign(mb.qd, state[19:])
    assign(mb.tau, action)
    
    pd.forward_kinematics(mb, mb.q, mb.qd)
    pd.forward_dynamics(mb, g)  # update qdd
    pd.integrate_euler_qdd(mb, ts)  # update qd, zero qdd
    dispatcher = world.get_collision_dispatcher()
    contacts = world.compute_contacts_multi_body([plane_mb, mb], dispatcher)[0]
    # we get 100 contacts: 
    #   4:   1 for each sphere 
    #   96:  8 for each box
    collisions = mb_solver.resolve_collision(contacts, ts)
    pd.integrate_euler(mb, ts)
    q_out = [mb.q[i] for i in range(mb.q.size())]
    qd_out = [mb.qd[i] for i in range(mb.qd.size())]
    return [*q_out, *qd_out]

def get_value(state, action):
    assign(mb.q, state[:19])
    assign(mb.qd, state[19:])
    assign(mb.tau, action)
    pd.forward_kinematics(mb, mb.q, mb.qd)
    pd.forward_dynamics(mb, g)  # update qdd
    qdd_out = [mb.qdd[i] for i in range(mb.qdd.size())]
    
    pd.integrate_euler_qdd(mb, ts)  # update qd, zero qdd
    dispatcher = world.get_collision_dispatcher()
    contacts = world.compute_contacts_multi_body([plane_mb, mb], dispatcher)[0]
    collisions = mb_solver.resolve_collision(contacts, ts)
    
    pd.integrate_euler(mb, ts)

    qd_out = [mb.qd[i] for i in range(mb.qd.size())]
    return [*qd_out, *qdd_out]    

def get_jacobian(function, argument):
    jacobian = []
    for i, x in enumerate(argument):
        argument[i] = pd.TinyDualDouble(x, 1.0)
        output = function(argument)
        jacobian.append([o.dual() for o in output])
        argument[i] = x
    return jacobian
```

```python
from matplotlib import pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (15, 10)
def plot_forces(forces):
    forces = np.array(forces)
    titles = ['Fx', 'Fy','Fz']
    legend = [link_names[f][-7:] for f in feet] + ['sum']
    for i in range(3):
        plt.subplot(3,1,i+1)
        plt.plot(forces[:,:,i])
    #     plt.plot(np.sum(forces[:,i:i+4], axis=1))
        plt.title(titles[i])
        plt.legend(legend)
        plt.grid()
        plt.show()
        

def tds_to_list(x):
    if dual:
        return [x[i].real() for i in range(x.size())]
    else:
        return [x[i] for i in range(x.size())]
```

```python
def simulate(n_iter, rtf=1.0):
    forces = np.zeros((n_iter, 4, 3))
    for j in range(n_iter):
        mb.tau.set_zero()
        joint_angles = tds_to_list(mb.q)[7:19]
        joint_velocities = tds_to_list(mb.qd)[6:18]
        for i in range(mb.tau.size()):
            mb.tau[i] = (
                KP[i%3] * (joint_angles_ref[i] - joint_angles[i])
                - KD[i%3] * joint_velocities[i]
            )
        pd.forward_kinematics(mb, mb.q, mb.qd)
        pd.forward_dynamics(mb, g)  # update qdd
        pd.integrate_euler_qdd(mb, ts)  # update qd, zero qdd
        multi_bodies = [plane_mb, mb]
        dispatcher = world.get_collision_dispatcher()
        contacts = world.compute_contacts_multi_body(multi_bodies, dispatcher)[0]
        # we get 100 contacts: 
        #   4:   1 for each sphere 
        #   96:  8 for each box
        before = [mb.qd[i] for i in range(mb.qd.size())]
        collisions = mb_solver.resolve_collision(contacts, ts)
        pd.integrate_euler(mb, ts)
        for c in collisions:
            i = feet.index(c.link_b)
            forces[j][i] = [ 
                c.lateral_friction_force_1.real(),
                c.lateral_friction_force_2.real(),
                c.normal_force.real()
            ]
#         [before[i].real() - mb.qd[i].real() for i in range(mb.qd.size())])
        # ncontacts = len(collisions)
        publish_joints(tds_to_list(mb.q))
        sleep(ts.real() / rtf)
    return forces
```
